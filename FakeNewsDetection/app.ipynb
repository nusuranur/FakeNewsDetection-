```python
fake['class'] = 0  # 0 for fake news
true['class'] = 1  # 1 for true news
```

```python
data = pd.concat([fake, true], axis=0)
```

```python
data.sample(10)
```

```python
data = data.drop(["title","subject","date"], axis=1)
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(["index"], axis=1, inplace=True)
```

```python
data.sample(5)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\[.*?\]", "", text)
    text = re.sub(r"\\W", " ", text)
    text = re.sub(r"https?://\S+|www\.\S+", "", text)
    text = re.sub(r"<.*?>+", "", text)
    text = re.sub(r"[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub(r"\n", "", text)
    text = re.sub(r"\w*\d\w*", "", text)
    return text
```

```python
data["text"] = data["text"].apply(clean_text)
```

```python
x=data["text"]
y=data["class"]

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
vectorsizer = TfidfVectorizer()
xtrain = vectorsizer.fit_transform(xtrain)
xv_test = vectorsizer.transform(xtest)
```

```python
lr = LogisticRegression()
lr.fit(xtrain, ytrain)
```

```python
predictions = lr.predict(xv_test)
lr.score(xv_test, ytest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, predictions))
```

```python
joblib.dump(vectorsizer, "vectorsizer.jb")
joblib.dump(lr, "lr_model.jb")
```
